Now we've filtered all the data, and we are only working with EUR.

We need to download the data on different time intervals. This way we can see which is better for the predictions

The creation for each dataset is way to long. For this reason, we are going to save all the data into diferent CSV files. This will reduce the amount of time needed for future executions.

This step is going to some hours, due to the amount of data we are generating. Just be pacient, and try to work on other projects while this is running.

```python
def create_file(currency, granularity):
    crypto_name = str(currency) + '-' + str(granularity)
    print("[STATUS] Creating {}...\n".format(crypto_name))

    file_name = './Datasets/{}/'.format(currency) + crypto_name + '.csv'
    dataframe = HistoricalData(str(currency), granularity, '2022-01-01-00-00', verbose=False).retrieve_data()
    dataframe.to_csv(file_name, index=True)

    print("[STATUS] {} created...\n".format(crypto_name))
    

def get_data(currency):
    if not os.path.exists('./Datasets/{}'.format(currency)):
        os.makedirs('./Datasets/{}'.format(currency))

    granularity_list = [60, 300, 900, 3600, 21600, 86400]

    pool = multiprocessing.Pool()

    # Use multiprocessing to create files in parallel
    pool.starmap(create_file, [(currency, granularity) for granularity in granularity_list])

    pool.close()
    pool.join()
```

```python
get_data('BTC-EUR')
```

```python
print("[STATUS] Convertion to csv completed.")
```
